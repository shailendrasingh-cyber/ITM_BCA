```python
# What are Conditional Statements?
# Conditional statements are used to perform different actions based on different conditions.

# For example:

# â€œIf itâ€™s raining, take an umbrella. Else, go out freely.â€

# In Python, we use if, elif, and else to implement such decisions.
```

```python
# Syntax of if-elif-else
```

```python
# The if Statement
age = 18
if age >= 18:
    print("You're eligible to vote.")

# Explanation:
# Here, age >= 18 is the condition. If it's True, the print statement is executed.
```

```python
 # The elif Statement
# Stands for "else if". Used to test multiple conditions.


marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")

# Flow:
# Python checks from top to bottom â†’ first True condition executes â†’ rest are ignored.
```

```python
# The else Statement
# Catches everything else that doesnâ€™t match any if or elif.


is_raining = False

if is_raining:
    print("Take an umbrella")
else:
    print("Enjoy the sunshine!")
```

```python
#  # Important Notes

# | Point                       | Explanation                                                                      |
# | --------------------------- | -------------------------------------------------------------------------------- |
# | **Indentation**             | Python uses indentation (tabs/spaces) to define code blocks. No `{}` like C/C++. |
# | **Boolean Expressions**     | Conditions must evaluate to `True` or `False`.                                   |
# | **You can nest conditions** | You can use `if` inside `if`.                                                    |

```

```python
#Nested Condition


age = 20
has_voter_id = True

if age >= 18:
    if has_voter_id:
        print("You can vote.")
    else:
        print("Apply for voter ID first.")
else:
    print("You are too young to vote.")
```

```python
# Python Calculator using if-elif-else

print("Simple Python Calculator")
print("Choose operation:")
print("1. Addition (+)")
print("2. Subtraction (-)")
print("3. Multiplication (*)")
print("4. Division (/)")
print("5. Modulus (%)")

# Take input from user
operation = input("Enter operation ( +, -, *, /, % ): ")
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Apply conditions
if operation == "+":
    result = num1 + num2
    print(f"Result: {num1} + {num2} = {result}")

elif operation == "-":
    result = num1 - num2
    print(f"Result: {num1} - {num2} = {result}")

elif operation == "*":
    result = num1 * num2
    print(f"Result: {num1} * {num2} = {result}")

elif operation == "/":
    if num2 != 0:
        result = num1 / num2
        print(f"Result: {num1} / {num2} = {result}")
    else:
        print("Error: Cannot divide by zero!")

elif operation == "%":
    result = num1 % num2
    print(f"Result: {num1} % {num2} = {result}")

else:
    print("Invalid operation! Please choose +, -, *, / or %")
```

```python
# Palindrome Checker (String Reversal)

text =  input("enter the text")

reversed_text =  text[::-1]

if text==reversed_text:
    print(f"yes your {text} is a palindrome")
else:
    print(f"the text you entered {text} is not palindrome")
```

```python
# Largest of Three Numbers

a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
c = float(input("Enter third number: "))

if a >= b and a >= c:
    print(f"The largest number is {a}")
elif b >= a and b >= c:
    print(f"The largest number is {b}")
else:
    print(f"The largest number is {c}")
```
