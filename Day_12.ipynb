```python
# 3. Function with Return Statement

# Concept:
# Functions can return values to the caller using return.

def area_of_circle(radius):
    pi =  3.14
    return  pi * radius*radius

#taking input
radi =  float(input("Enter the radius"))

print("Area of circle:-" ,area_of_circle(radi) , "cmsq")
```

```python
# another example of the function with return 
def square_of_number(side):
    return side*side

inp =  int(input("enter the value of the side"))
print("Area of the square:-" , square_of_number(inp))
```

```python
# 4. Default Arguments
# Concept:

# You can assign default values to parameters.
# If no argument is passed, the default value is used.

def greet_1(name =  "Student"):
    print("Hello Mr." , name)

greet_1()
# But after we pass the argument
greet_1("Aakash")
```

```python
# 5. Keyword Arguments
# Concept:

# You can pass arguments using parameter names.

# Example:
def sample_2(name  , age):
    print("Name of the student is " , name)
    print("Age of the student is " , age)

sample_2("aakash" , 23)
```

```python
# 6. Variable-Length Arguments
# Concept:

# When you donâ€™t know how many arguments will be passed:

# *args for non-keyword arguments (tuple)

# **kwargs for keyword arguments (dictionary)

# example of the *args
def sample_3(*marks):
    print("Marks are " , marks)
sample_3(10 , 20 , 20  , "demo" , True )

# Example of the **kwargs

def sample_3a(**info):
    print("Information enter in this function are" , info)

sample_3a(name="Harish",Age=30 )
```

```python
# 7. Scope of Variables (Local vs Global)
# The Variable use inside the funtion is called the local 
# Outside of the fucntion is known as global 

x =  20
def demo_1():
    x =  30
    print("Inside fucntion" , x)

demo_1()
print("The global Variable " , x)
```

```python
# 8. Lambda (Anonymous) Functions
# Concept:

# Small, one-line functions without a name.

# Example:

demo_3 =  lambda y:y*y
print("this is the demo correct use of the lambda  function" , demo_3(3))

# For two positional Argument

add_two_number  =  lambda a , b :a+b
print(add_two_number(2 , 3))

#  Find the max of the in the number

mac_num =  lambda a , b : a if a>b else b 
print(mac_num(2 , 3))
```

```python
# 9. Functions Returning Multiple Values
# Example:
def multiple_function (a , b):
    add =  a+b
    sub =  a-b
    multiply =  a*b
    return add ,sub, multiply

result =  multiple_function(1 , 2)
print(result)

```

```python
# 10. Nested Functions
# Concept:

# A function inside another function.

def outer_function():
    def inner_function():
        print("this is the output of the inner function")
    print("this is the output of the outer function")
    inner_function()
outer_function()
```

```python
def outer_1(a, b):
    def inner_1(a , b ):
        print("the output of the inner function is" , a+b)
    print("the output of the outer function is" , a-b)
    inner_1(1 , 2)

outer_1(1 , 2)
    
```

```python
# 11. Recursive Functions
# Concept:

# A function that calls itself.

# Example:

def factorial(x):
    if x ==1:
        return 1
    else:
        return x* factorial(x-1)

print("factorial of", factorial(5))
```

```python
# sum of the first natural number

def  natural_num(n):
    if n == 0:
        return 0
    else:
        return n + natural_num(n-1)

usr_in =  int(input("Enter the number"))
print("The sum 0f the first natural number is" , natural_num(usr_in))
```
