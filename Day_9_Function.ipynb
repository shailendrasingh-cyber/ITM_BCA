```python
 # Variable-Length Arguments
 # *args (Non-keyword arguments as tuple)("it does not see the len and what ever inside of the number ti store in tuple and add all ")

def add_all(*args):
    return sum(args)

print(add_all(1 ,2 ,3 ,4))
```

```python
 # **kwargs (Keyword arguments as dictionary)

def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Vikas", age=23, city="Delhi")
```

```python
# 8. Scope of Variables
# Local: Inside function

# # Global: Outside all functions


x = 10  # global

def show():
    x = 5  # local
    print("Inside:", x)

show()
print("Outside:", x)
```

```python
add = lambda x, y: x + y
print(add(3, 4))

# A lambda function to add 10 to a number
add_ten = lambda x: x + 10
print(add_ten(5))  # Output: 15

# A lambda function used directly as an argument to a higher-order function
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x * x, numbers))
print(squared_numbers)  # Output: [1, 4, 9, 16, 25]
```

```python
# 11. Function as Argument
def shout(text):
    return text.upper()

def whisper(text):
    return text.lower()

def speak(func):
    print(func("Hello Baby"))

speak(shout)
speak(whisper)

```

```python
# 1. Function to Find Factorial of a Number

def factorial(n):
    result = 1
    for i in range(1 , n+1):
        result *= i
    return(result)

print(factorial(10))
```

```python
 # 2. Function That Takes a List and Returns Sum of Even Numbers

def even_number(number):
    total =  0
    for num in number:
        if num%2==0:
            total+=num
    return(total)

print(even_number([1, 2 ,3 ,4 ,5 ,6]))
```

```python
# 3. Function to Check if a Number is Prime


def is_prime_while(n):
    if n <= 1:
        return False
    divisor = 2
    while divisor * divisor <= n: # Same condition as i <= int(n**0.5)
        if n % divisor == 0:
            return False
        divisor += 1 # Increment the divisor
    return True

print(is_prime_while(5))  # Output: True
print(is_prime_while(10)) # Output: False
```

```python
# 5. Function to Reverse a String

def reverse_string(s):
    s =  s[::-1]
    return s

print(reverse_string("hello"))
```
