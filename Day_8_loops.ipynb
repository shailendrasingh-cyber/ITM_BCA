```python
# WHY Use Loops?
# Imagine you have to:

# Print "Hello" 100 times.
# You could write print("Hello") 100 times…
# OR use a loop and do it in 2 lines.
```

```python
# # Types of Loops in Python

# | Loop Type  | When to Use                              |
# | ---------- | ---------------------------------------- |
# | `for`      | Repeating over **known values** or range |
# | `while`    | Repeating based on **a condition**       |
# | `nested`   | Loops inside loops (grids, patterns)     |
# | `break`    | To stop loop early                       |
# | `continue` | To skip one iteration                    |
# | `else`     | To run code if loop ends cleanly         |
```

```python
# # 1. FOR LOOP
# # Syntax:
# # for variable in iterable:
# #     # block of code


# iterable = list, string, tuple, range, etc.

# The loop runs once for each value in the iterable.
```

```python
# Example 1: Print numbers 0 to 4

for i in range(5):  # i = 0 to 4
    print("i =", i)
```

```python
# Example 2: Loop through a list
colors = ['red', 'blue', 'green']
for color in colors:
    print(color)
```

```python
# Example 3: Reverse Loop

for i in range(10, 0, -1):
    print(i)
```

```python
# -------------------------------2. WHILE LOOP ----------------------------####
```

```python
# Syntax:

# while condition:
#     # block of code
```

```python
 # Example: Print 1 to 5

num =  1
while num <=5:
    print(num)
    num+=1
```

```python
 # 3. Loop Control Statements
for i in range(10):
    if i == 7:
        break
    print(i)
```

```python
# continue → Skip current loop iteration

for i in range(6):
    if i == 6:
        continue
    print(i)
```

```python
 # pass → Placeholder, does nothing

for i in range(10):
    if i == 5:
        pass
    print(i)
```

```python
 # 4. FOR-ELSE and WHILE-ELSE

for i in range(10):
    print(i)
else:
    print("Loop is finished")
```

```python
# BUT   if we have to use the break usinf else in the for loop 
for i in range(10):
    if i == 6:
        break
    print(i)
else:
    print("loop finished sucessfully") # this line won't run 

 # The else runs only if loop is not broken by break
```

```python
# 1. Right-Angled Triangle (Left-Aligned)
row =  5 
for i in range (1 , row +1):
    print("  *  "  * i)
```

```python
#full traingle
rows = 5

for i in range(1, rows + 1):
    spaces = "  " * (rows - i)
    stars = " *" * (2 * i - 1)
    print(spaces + stars)
```

```python
# left side  right angle traingle
row  =  5 
for i in range (1  , row+1):
    print(" "*(row-i) + "*"*i)
```

```python
row =  5
for i in range(row , 0 , -1):
    print(" "* (row-i)+"*"*i)
```

```python
 # Pyramid Pattern (Center-Aligned)
row  =  5 
for i in range(1 , row+1):
    print(" " * (row-i) + "*"*(2*i-1))
```

```python
# inverted pyramid 
row = 5 
for i in range(row , 0 , -1 ):
    print(" " *(row-i) + "*"*(2*i-1))
```

```python
row = 10
for i in range(row):
    for j in range(row - i - 1):
        print(" ", end="")
    for j in range(2 * i + 1):
        print("*", end="")
    print()
```

```python
#define in1 9
#define in2 8
#define in3 7
#define in4 6

void setup() {
  Serial.begin(9600);
  pinMode(in1, OUTPUT);
  pinMode(in2, OUTPUT);
  pinMode(in3, OUTPUT);
  pinMode(in4, OUTPUT);
}

void loop() {
  if (Serial.available() > 0) {
    char inputvalue = char(Serial.read());

    if (inputvalue == 'F') {
      forward();
    } 
    else if (inputvalue == 'B') {
      backward();
    } 
    else if (inputvalue == 'L') {
      left();
    } 
    else if (inputvalue == 'R') {
      right();
    } 
    else if (inputvalue == 'S') {
      stopMotors();
    }
  }
}

// Function to move forward
void forward() {
  digitalWrite(in1, HIGH);
  digitalWrite(in2, LOW);
  digitalWrite(in3, HIGH);
  digitalWrite(in4, LOW);
}

// Function to move backward
void backward() {
  digitalWrite(in1, LOW);
  digitalWrite(in2, HIGH);
  digitalWrite(in3, LOW);
  digitalWrite(in4, HIGH);
}

// Function to turn left
void left() {
  digitalWrite(in1, HIGH);
  digitalWrite(in2, LOW);
  digitalWrite(in3, LOW);
  digitalWrite(in4, LOW);
}

// Function to turn right
void right() {
  digitalWrite(in1, LOW);
  digitalWrite(in2, LOW);
  digitalWrite(in3, HIGH);
  digitalWrite(in4, LOW);
}

// Function to stop
void stopMotors() {
  digitalWrite(in1, LOW);
  digitalWrite(in2, LOW);
  digitalWrite(in3, LOW);
  digitalWrite(in4, LOW);
}
```
