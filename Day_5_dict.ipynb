```python
# What is a Dictionary in Python?
# A dictionary in Python is a collection of key-value pairs.
# It is unordered, mutable, and indexed by keys, not positions.
```

```python
# Sample
my_dict = {
    "key1": "value1",
    "key2": "value2"
}
print(my_dict)
```

```python
# Creating a dictionary
student = {
    "name": "Alice",
    "age": 21,
    "course": "Computer Science"
}
print("Student Dictionary:", student)
```

```python
# Accessing values (using Keys)
print("Name:", student["name"])
print("Age:", student["age"])
```

```python
# 3. Adding a New Key-Value Pair

# Adding a new item
student["grade"] = "A"
print("After Adding Grade:", student)
```

```python
# 4. Updating an Existing Keyâ€™s Value

# Updating a value
student["age"] = 22
print("After Updating Age:", student)
```

```python
# 5. Removing Items

# Remove item using del
del student["grade"]
print("After Deleting Grade:", student)

# Using pop()
course = student.pop("course")
print("Popped Course:", course)
print("After Popping Course:", student)
```

```python
# 6 Using get() to Avoid Error
```

```python
# Using get() method
print("Get Name:", student.get("name"))
print("Get Grade (not present):", student.get("grade", "Not Found"))
```

```python
 # 7. Looping Through Dictionary

# Keys
print("Keys:")
for key in student:
    print(key)

# Values
print("Values:")
for value in student.values():
    print(value)

# Key-Value Pairs
print("Key-Value Pairs:")
for key, value in student.items():
    print(key, ":", value)
```

```python
# 8. Length of Dictionary
print("Total Items:", len(student))

```

```python
# 9. Clear All Items
# Dictionary inside a dictionary
students = {
    "student1": {"name": "Alice", "age": 21},
    "student2": {"name": "Bob", "age": 22}
}
print("Nested Dictionary:", students)
print("Student1 Name:", students["student1"]["name"])

```

```python
 # Summary Points:

# | Operation  | Method/Action              |
# | ---------- | -------------------------- |
# | Create     | `{}` or `dict()`           |
# | Access     | `dict[key]` or `get()`     |
# | Add/Update | `dict[key] = value`        |
# | Delete     | `del`, `pop()`, `clear()`  |
# | Loop       | `for`, `items()`, `keys()` |
# | Nested     | Dict inside dict           |
```
