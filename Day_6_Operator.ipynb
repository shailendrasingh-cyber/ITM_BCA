```python
# Operator :- 
# Operators in Python are special symbols used to perform operations on variables and values.
```

```python
# #  1. Arithmetic Operators
# # Used to perform basic mathematical operations.


# | Operator | Description         | Example   | Output |
# | -------- | ------------------- | --------- | ------ |
# | `+`      | Addition            | `10 + 5`  | `15`   |
# | `-`      | Subtraction         | `10 - 5`  | `5`    |
# | `*`      | Multiplication      | `10 * 5`  | `50`   |
# | `/`      | Division            | `10 / 5`  | `2.0`  |
# | `//`     | Floor Division      | `10 // 3` | `3`    |
# | `%`      | Modulus (remainder) | `10 % 3`  | `1`    |
# | `**`     | Exponentiation      | `2 ** 3`  | `8`    |

```

```python
# # 2. Comparison Operators (Relational)
# # Used to compare two values. The result is always True or False.

# | Operator | Description      | Example  | Output  |
# | -------- | ---------------- | -------- | ------- |
# | `==`     | Equal            | `5 == 5` | `True`  |
# | `!=`     | Not Equal        | `5 != 3` | `True`  |
# | `>`      | Greater Than     | `5 > 3`  | `True`  |
# | `<`      | Less Than        | `5 < 3`  | `False` |
# | `>=`     | Greater or Equal | `5 >= 5` | `True`  |
# | `<=`     | Less or Equal    | `3 <= 4` | `True`  |
```

```python
# 3. Logical Operators
# Used to combine conditional statements.

# | Operator | Description                  | Example            | Output  |
# | -------- | ---------------------------- | ------------------ | ------- |
# | `and`    | True if both are True        | `5 > 2 and 5 < 10` | `True`  |
# | `or`     | True if at least one is True | `5 < 2 or 5 < 10`  | `True`  |
# | `not`    | Reverses the result          | `not(5 > 2)`       | `False` |


```

```python
# 4. Assignment Operators
# Used to assign values to variables, with support for shorthand operations.

# | Operator | Description             | Example   | Meaning      |
# | -------- | ----------------------- | --------- | ------------ |
# | `=`      | Assign                  | `x = 5`   | x = 5        |
# | `+=`     | Add and assign          | `x += 2`  | x = x + 2    |
# | `-=`     | Subtract and assign     | `x -= 3`  | x = x - 3    |
# | `*=`     | Multiply and assign     | `x *= 4`  | x = x \* 4   |
# | `/=`     | Divide and assign       | `x /= 2`  | x = x / 2    |
# | `//=`    | Floor divide and assign | `x //= 2` | x = x // 2   |
# | `%=`     | Modulo and assign       | `x %= 3`  | x = x % 3    |
# | `**=`    | Power and assign        | `x **= 2` | x = x \*\* 2 |
```

```python
# # 5. Bitwise Operators
# # Work on bits and perform bit-by-bit operations.

# | Operator | Description | Example  | Binary Example                    | Output |        |               |     |
# | -------- | ----------- | -------- | --------------------------------- | ------ | ------ | ------------- | --- |
# | `&`      | AND         | `5 & 3`  | `0101 & 0011` → `0001`            | `1`    |        |               |     |
# | \`       | \`          | OR       | \`5                               | 3\`    | \`0101 | 0011`→`0111\` | `7` |
# | `^`      | XOR         | `5 ^ 3`  | `0101 ^ 0011` → `0110`            | `6`    |        |               |     |
# | `~`      | NOT         | `~5`     | `~0101` → `1010` (2's complement) | `-6`   |        |               |     |
# | `<<`     | Left Shift  | `5 << 1` | `0101` → `1010`                   | `10`   |        |               |     |
# | `>>`     | Right Shift | `5 >> 1` | `0101` → `0010`                   | `2`    |        |               |     |

```

```python
# # 6. Identity Operators
# # Used to compare memory locations of two objects.

# | Operator | Description             | Example      | Output     |
# | -------- | ----------------------- | ------------ | ---------- |
# | `is`     | True if same object     | `a is b`     | True/False |
# | `is not` | True if not same object | `a is not b` | True/False |

```

```python
# 7. Membership Operators
# Used to test if a sequence contains a value.

# Operator	Description	Example	Output
# in	True if present	2 in [1, 2, 3]	True
# not in	True if not present	'x' not in 'abc'	True

```

```python
# # Summary Table


# | Category   | Examples                 |                 |
# | ---------- | ------------------------ | --------------- |
# | Arithmetic | +, -, \*, /, //, %, \*\* |                 |
# | Comparison | ==, !=, >, <, >=, <=     |                 |
# | Logical    | and, or, not             |                 |
# | Assignment | =, +=, -=, \*=, /=, etc. |                 |
# | Bitwise    | &,                       | , ^, \~, <<, >> |
# | Identity   | is, is not               |                 |
# | Membership | in, not in               |                 |

```

```python
 # Example Program Using All Python Operators

# 1. Arithmetic Operators
a = 10
b = 3

print("Arithmetic Operators:")
print("Addition:", a + b)         # 13
print("Subtraction:", a - b)      # 7
print("Multiplication:", a * b)   # 30
print("Division:", a / b)         # 3.333...
print("Floor Division:", a // b)  # 3
print("Modulus:", a % b)          # 1
print("Exponent:", a ** b)        # 1000

print("\nComparison Operators:")
# 2. Comparison Operators
print("a == b:", a == b)          # False
print("a != b:", a != b)          # True
print("a > b:", a > b)            # True
print("a < b:", a < b)            # False
print("a >= b:", a >= b)          # True
print("a <= b:", a <= b)          # False

print("\nLogical Operators:")
# 3. Logical Operators
x = True
y = False
print("x and y:", x and y)        # False
print("x or y:", x or y)          # True
print("not x:", not x)            # False

print("\nAssignment Operators:")
# 4. Assignment Operators
c = 5
print("Initial c:", c)
c += 2
print("After c += 2:", c)         # 7
c *= 3
print("After c *= 3:", c)         # 21
c -= 4
print("After c -= 4:", c)         # 17
c //= 2
print("After c //= 2:", c)        # 8
c %= 5
print("After c %= 5:", c)         # 3
c **= 2
print("After c **= 2:", c)        # 9

print("\nBitwise Operators:")
# 5. Bitwise Operators
p = 5     # 0101
q = 3     # 0011
print("p & q:", p & q)            # 1
print("p | q:", p | q)            # 7
print("p ^ q:", p ^ q)            # 6
print("~p:", ~p)                  # -6 (2's complement)
print("p << 1:", p << 1)          # 10 (left shift)
print("p >> 1:", p >> 1)          # 2  (right shift)

print("\nIdentity Operators:")
# 6. Identity Operators
list1 = [1, 2]
list2 = [1, 2]
list3 = list1

print("list1 is list2:", list1 is list2)    # False
print("list1 is list3:", list1 is list3)    # True
print("list1 is not list2:", list1 is not list2)  # True

print("\nMembership Operators:")
# 7. Membership Operators
colors = ['red', 'blue', 'green']
print("'red' in colors:", 'red' in colors)            # True
print("'yellow' not in colors:", 'yellow' not in colors)  # True
```
