```python
# # __init__() Constructor Method
# # This special method is used to initialize object data when an object is created.

#  Explanation:

# __init__() is a constructor. It runs automatically when an object is created.

# self is the object itself. Think of it as this in other languages.

# brand and model are parameters passed during object creation.

# self.brand = brand means storing the parameter brand into object’s own variable brand.

class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model


car1 = Car("BMW", "X5")
print(car1.brand)  # BMW
print(car1.model)  # X5
```

```python
# Class Attribute vs Object Attribute

class college:
    colleges = "abc college"

    def __init__(self , name):
        self.name =  name
s1 =  college("manish")
s2 =  college("anish")

print(s1.name)
print(s2.name)
print(s1.colleges)
print(s2.colleges)



# college is shared by all students.

# name is different for each student object.
```

```python
class college:
    colleges = "abc college"

    def __init__(self, name):
        self.name = name

    def info(self):
        print(f"The name of the student is {self.name} and he studies in {self.colleges}")

s1 = college("manish")
s2 = college("anish")

s1.info()
s2.info()
```

```python
# Creating Your Own Method (Function inside class)
# You can define behavior using methods.

class car:
    def __init__(self  , brand):
        self.brand = brand

    def demo(self):
        print(f"{self.brand} is driving")


car_1 =  car("ford")
car_1.demo()
```

```python
# Using __str__() to Beautify Print Output

class fruit:
    def __init__(self   , name):
        self.name =  name
    def __str__(self):
        return f"fruit name is {self.name}"

f =  fruit("mango")
print(f)

# Without __str__(), printing the object shows memory location.

# With __str__(), it shows a pretty message.
```

```python
 # Encapsulation (Protecting Data)

class bankaccount:
    def __init__(self , balance):
        self.balance =  balance

    def deposit(self , amount):
        self.balance+=amount
    def show_balance(self):
        print(f"the toatal balance in your account is {self.balance}")

acc =  bankaccount(1000)
acc.deposit(2000)
acc.show_balance()


# __balance is private. You can't access it directly.

# Only class methods can change or show it.
```

```python
class bankaccount:
    def __init__(self, balance):
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance. Withdrawal failed.")
        else:
            self.balance -= amount
            print(f"You have withdrawn {amount}. Remaining balance: {self.balance}")

    def show_balance(self):
        print(f"The total balance in your account is {self.balance}")

acc = bankaccount(1000)
acc.deposit(2000)
acc.show_balance()

acc.withdraw(500)       # Valid withdrawal
acc.withdraw(5000)      # Invalid withdrawal
acc.show_balance()
```

```python
 # Inheritance (Getting features from another class)

class animal:
    def  sound(self):
        print("some random sound")

class dog(animal):
    def sound(self):
        print("bark")

c=  dog()
c.sound()

# Dog inherits Animal using (Animal).

# Dog can override the method with its own version.
```

```python
 # Using super() to Call Parent Constructor

class Animal:
    def __init__(self):
        print("Animal created")

class Dog(Animal):
    def __init__(self):
        super().__init__()  # Calls Animal's __init__()
        print("Dog created")

d = Dog()
```

```python
 # Polymorphism (Same method name, different behavior)

class Cat:
    def speak(self):
        print("Meow")

class Bird:
    def speak(self):
        print("Chirp")

def make_sound(animal):
    animal.speak()

make_sound(Cat()) 
make_sound(Bird()) 

# speak() is defined differently in each class.

# make_sound() works for both.
```
